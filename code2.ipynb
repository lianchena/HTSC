```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# 
start_date = "2010-01-01"
end_date = "2020-12-31"
index_symbol = "^DJI"  # 500A

# yfinance
index_data = yf.download(index_symbol, start=start_date, end=end_date)

# 101108
national_day_start = "2020-10-01"
national_day_end = "2020-10-08"

# 
pre_national_day_data = index_data[:national_day_start]
post_national_day_data = index_data[national_day_end:]

# 
pre_national_day_return = (pre_national_day_data["Adj Close"] / pre_national_day_data["Adj Close"].iloc[0] - 1) * 100
post_national_day_return = (post_national_day_data["Adj Close"] / post_national_day_data["Adj Close"].iloc[0] - 1) * 100

# 
print("Before:", pre_national_day_return.iloc[-1], "%")
print("After:", post_national_day_return.iloc[-1], "%")

# 
plt.figure(figsize=(10, 6))
plt.plot(pre_national_day_return, label="Before", linestyle="-", marker="o")
plt.plot(post_national_day_return, label="After", linestyle="-", marker="o")
plt.xlabel("date")
plt.ylabel("cum_ret (%)")
plt.title("National_Day")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame
data = pd.DataFrame({
    "Date": ["2021-09-25", "2021-09-26", "2021-09-27", "2021-09-28", "2021-09-29", "2021-10-01", "2021-10-02", "2021-10-03"],
    "Market_Return": [-0.5, 0.2, -0.1, 0.4, -0.3, 0.5, 0.3, 0.6]
})

# 
data["Date"] = pd.to_datetime(data["Date"])

# 
national_day_start = "2021-10-01"
pre_national_day_data = data[data["Date"] < national_day_start]
post_national_day_data = data[data["Date"] >= national_day_start]

# 
pre_national_day_mean_return = pre_national_day_data["Market_Return"].mean()
post_national_day_mean_return = post_national_day_data["Market_Return"].mean()

# 
print("Before:", pre_national_day_mean_return)
print("After:", post_national_day_mean_return)

# 
plt.figure(figsize=(8, 6))
plt.bar(["Before", "After"], [pre_national_day_mean_return, post_national_day_mean_return], color=['red', 'green'])
plt.xlabel("B_A")
plt.ylabel("avg_ret")
plt.title("National_Day")
plt.show()
```
